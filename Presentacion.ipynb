### 3. Elasticidad Precio-Demanda

La elasticidad-precio de la demanda
Mide la variacin porcentual que experimenta la cantidad demandada como consecuencia de una variacin del precio de un 1%

Interpretacin:
Si la demanda es muy sensible al precio (muy elstica) una subida del precio reduce la demanda tanto que disminuye el ingreso total. En cambio, si la demanda es muy poco sensible al precio (muy inelstica) una subida del precio no altera mucho la demanda y aumenta el ingreso total.

La frmula que utilizariamos para calcular la elasticidad sera la siguiente:

p = (1/1)*(p1/x1)

Si p > 1: Tendremos una demanda elstica,  la cantidad demandada cambia ms que proporcionalmente con respecto al precio.

Si p < 1: Tendremos una demanda inelstica, la cantidad demandada responde menos que proporcionalmente al cambio en el precio.

Si p = 1: La demanda es unitaria, lo cual significa que la cantidad demandada cambia en la misma proporcin que el cambio en el precio.
La problemtica que tenemos en este caso, es que no tenemos funcin de demanda y no podemos hacer una derivada parcial para encontrar la cantidad demandada de x1 respecto a p1. Por ello analizamos el problema de manera discreta y calculando la elasticidad de la demanda con la frmula siguiente: p = (Q/Q)/(P/P)

Siendo:

Q: Cambio en la cantidad demandada entre dos puntos en el tiempo (aos en este caso).

P: Cambio en el precio medio entre dos aos consecutivos.

P y Q: Precio medio y cantidad demandada de un ao especfico.
